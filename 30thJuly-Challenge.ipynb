#### 1.Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave

```python
apr['converted_date'] = pd.to_datetime(apr['Date'])
apr['week'] = apr['converted_date'].dt.isocalendar().week
apr[apr['Total Chats'] != 0].groupby(['Agent Name','week'])['Average Rating'].mean()
```

#### 9.average weekly response time for each agent

```python
apr['avg_response'] = pd.to_timedelta(apr['Average Response Time'])
apr['converted_date'] = pd.to_datetime(apr['Date'])
apr['week'] = apr['converted_date'].dt.isocalendar().week
apr[apr['Total Chats'] != 0].groupby(['Agent Name','week'])['avg_response'].mean()
```

#### 10.Average weekly resolution time for each agents

```python
apr['avg_resolution'] = pd.to_timedelta(apr['Average Resolution Time'])
apr['converted_date'] = pd.to_datetime(apr['Date'])
apr['week'] = apr['converted_date'].dt.isocalendar().week
apr[apr['Total Chats'] != 0].groupby(['Agent Name','week'])['avg_resolution'].mean()
```
